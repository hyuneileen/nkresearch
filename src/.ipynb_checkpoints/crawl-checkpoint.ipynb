```python
import requests
from bs4 import BeautifulSoup

import progressbar
from time import sleep
bar = progressbar.ProgressBar(maxval=351, \
    widgets=[progressbar.Bar('=', '[', ']'), ' ', progressbar.Percentage()])

bar.start()
collection = []
for i in list(range(1,351)):
    soup = BeautifulSoup(requests.get("http://www.ryongnamsan.edu.kp/univ/ko/research/journals?cp="+str(i)).content, "html.parser")
    for d in soup.find_all("a", {"class": "journal-title"},href=True):
        title = d.text.strip()
        author = d.findNext("div", {"class": "list-author"}).text.strip()
        url = d['href']
        collection.append([title,author,url])
        bar.update(i+1)
bar.finish()
```

```python
import json
with open('collection.json', 'w') as f:
    json.dump(collection, f,ensure_ascii=False,indent=2)
```

```python
import requests
from bs4 import BeautifulSoup
import json

def site_crawler():
    languages = ["ko","en"]
    collection = []
    ## [listing count, front page titles]
    current_listing_count = {}
    for lang in languages:
        ## getting digits
        soup = BeautifulSoup(requests.get("http://www.ryongnamsan.edu.kp/univ/"+lang+"/research/journals").content, "html.parser")
        total_listings = soup.find("input", {"id": "totalListCnt"}).get('value')
        total_listings = total_listings.replace(" ","")
        max_lim = round(int(total_listings)/17)
        
        ## get subjects
        ## used later as metadata
        idx_to_name = {}
        for i,tag in enumerate(soup.find_all('nobr')):
            if i not in [0,len(soup.find_all('nobr'))-1]:
                subject = tag.text
                v = tag.find('a')['href']
                idx = v.split('/')[-1].strip()
                idx_to_name[idx]= subject

        ## append with first page since it's already loaded
        ## also save first page & listing count for HTML comparison
        for d in soup.find_all("a", {"class": "journal-title"},href=True):
            title = d.text.strip()
            author = d.findNext("div", {"class": "list-author"}).text.strip()
            url = d['href']
            collection.append([title,author,url])

        ## beginning other pages...
        for i in list(range(1,max_lim)):
            soup = BeautifulSoup(requests.get("http://www.ryongnamsan.edu.kp/univ/"+lang+"/research/journals?cp="+str(i)).content, "html.parser")
            for d in soup.find_all("a", {"class": "journal-title"},href=True):
                title = d.text.strip()
                author = d.findNext("div", {"class": "list-author"}).text.strip()
                url = d['href']
                collection.append([title,author,url])
        
        ## get acutal sub listing counts
        for journalidx in idx_to_name:
            soup = BeautifulSoup(requests.get("http://www.ryongnamsan.edu.kp/univ/"+lang+"/research/journals/"+str(journalidx)).content, "html.parser")
            sub_listing_count = soup.find("input", {"id": "totalListCnt"}).get('value')
            sub_listing_count = sub_listing_count.replace(" ","")
            idx_to_name[journalidx] = [subject,sub_listing_count]
        
        ## get collected sub listing counts
        
        
        ## not all were fetched
        if len(collection) < len(total_listings):
            ## compare sub listing counts
            for journalidx in idx_to_name:
                current_sub_listings = [c[2] for c in collection if ]
        
    ## write to json
    with open('src/collection.json', 'w') as f:
        json.dump(collection, f, ensure_ascii=False, indent=2)
    with open('src/current_HTML.json', 'w') as f:
        json.dump(current_HTML, f, ensure_ascii=False, indent=2)
    
    return collection,current_HTML
```
