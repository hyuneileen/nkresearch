```python
import requests

url = "https://service.elsevier.com/app/answers/detail/a_id/15181/supporthub/scopus/"
```

```python
import lxml.html as lh
page = requests.get(url)
doc = lh.fromstring(page.content)
tr_elements = doc.xpath('//tr')
```

```python
tr_elements = doc.xpath('//tr')#Create empty list
col=[]
i=0#For each row, store each first element (header) and an empty list
for t in tr_elements[0]:
    i+=1
    name=t.text_content()
    print ('%d:"%s"'%(i,name))
    col.append((name,[]))
```

```python
#Since out first row is the header, data is stored on the second row onwards
for j in range(1,len(tr_elements)):
    #T is our j'th row
    T=tr_elements[j]
    
    #i is the index of our column
    i=0
    
    #Iterate through each element of the row
    for t in T.iterchildren():
        data=t.text_content() 
        #Check if row is empty
        if i>0:
        #Convert any numerical value to integers
            try:
                data=int(data)
            except:
                pass
        #Append the data to the empty list of the i'th column
        col[i][1].append(data)
        #Increment i for the next column
        i+=1
```

```python
import pandas as pd
data={title:column for (title,column) in col}
df=pd.DataFrame(data)
```

```python
ds = [asjc_codes1, asjc_codes2]
d = {}
for k in asjc_codes1.keys():
    d[int(k)] = tuple(d[k] for d in ds)
```

```python
for item in ref_db:
    if item["scopus"] and 'ASJC Codes' in item["scopus"]:
        asjc_fields = []
        for a in item["scopus"]['ASJC Codes']:
            field,area = d[a]
            asjc_fields.append(area)
        item["scopus"].update({"ASJC Fields":asjc_fields})
```
