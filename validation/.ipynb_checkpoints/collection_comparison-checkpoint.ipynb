# Part I: Crawling UniKorea's site

```python
import requests
from bs4 import BeautifulSoup
from itertools import groupby
```

```python
page1 = "https://unibook.unikorea.go.kr/material/list?materialScope=SER&nation=kn&format=image&method=&fields=ALL&keywords=%EA%B9%80%EC%9D%BC%EC%84%B1&conjunctions=AND&fields=ALL&keywords=%EA%B9%80%EC%9D%BC%EC%84%B1%EC%A2%85%ED%95%A9%EB%8C%80%ED%95%99+%ED%95%99%EB%B3%B4&pageSize=10&sortField=publishYear&sortDirection=DESCENDING"
page2 = "https://unibook.unikorea.go.kr/material/list?materialScope=SER&nation=kn&format=image&method=&fields=ALL&keywords=%EA%B9%80%EC%9D%BC%EC%84%B1&conjunctions=AND&fields=ALL&keywords=%EA%B9%80%EC%9D%BC%EC%84%B1%EC%A2%85%ED%95%A9%EB%8C%80%ED%95%99+%ED%95%99%EB%B3%B4&pageSize=10&sortField=publishYear&sortDirection=DESCENDING&page=2"
pages = [page1,page2]

# finding all individual journal listings to recursively fetch them
# formatted as: [subject name, listing link]
listings = []
for page in pages:
    soup = BeautifulSoup(requests.get(page).content, "html.parser")
    for div in soup.find_all("li", {"class": "item-img"}):
        for img in div.find_all('img', alt=True):
            title = img['alt'].strip()
        a = div.findNext('a', href=True)
        link = a["href"]
        listings.append([title,link])

# deleting first one since it's a copy
del listings[0]
print ("Including journal {} with link \n{}, \na total of {} journal listings were found.".format(listings[0][0],listings[0][1],len(listings)))
```

```python
base = "https://unibook.unikorea.go.kr/material/"
all_listings = []
# visiting the link of each listing found earlier
for listing in listings:
    # adding base domain link to href
    listing_link = base+listing[1]
    # link with "checkins" loads collection information
    listing_link = listing_link.replace("view","checkins",1)
    
    # grabbing the link
    soup = BeautifulSoup(requests.get(listing_link).content, "html.parser")
    
    # scrapes table with info on each volume of the journal available
    combined = []
    for td in soup.find_all("td",{"align": "left"}):
        table = td.text.strip()
        # table contains data: [journal volume, publication date, date seized, acquisition status, 
        #                      total article count, href to each volume's information on the articles]
        combined.append(table)
        # finding href to recursively crawl each volume
        # there is one volume for each year; 
        # there are numbers within each volume
        for aa in td.find_all("a",href=True):
            combined.append(aa["href"])
    
    article_crawl = []
    # restructuring data to group volumes together
    volumes = [list(g) for k, g in groupby(combined, key=bool) if k]
    
    for volume in volumes:
        href = volume[-1]
        # adding href to base url again
        # recursively fetching article info for each volume listed
        article_link = base+href
        soup = BeautifulSoup(requests.get(article_link).content, "html.parser")
        articles = []
        # scraping table that lists the title, author, and page range
        # of each article belonging to that volume
        
        # tr = each table row
        for tr in soup.find_all('tr'):
            tds = []
            # td = data in each cell
            # each cell contains info on one article
            for td in tr.find_all("td"):
                tds.append (td.text.strip())
            articles.append(tds)
        # removing empties
        articles = [a for a in articles if a]
        articles = [a for a in articles if len(a) == 5]
        
        # deleting index row
        del articles[0]
        article_crawl.append([volume,articles])
    all_listings.append(article_crawl)
    print ("Finished crawling journal {}.".format(listing[0]))
```

```python
subject_names = [l[0].split(": ")[1].strip() for l in listings]
print ("The available journals are:\n",", ".join(subject_names))
```

```python
# re-organizing crawled info into a dictionary
unikorea_crawl = dict(zip(subject_names, all_listings))

# cleaning data
import re
base = "https://unibook.unikorea.go.kr/material/"
for k,v in unikorea_crawl.items():
    for i,volume in enumerate(v):
        issue = re.findall(r'\d+', volume[0][0])
        issue = [d for d in issue if len(d) <= 2]
        vol,no = issue
        date_seized, status, collection_count = volume[0][2:5]
        link = base+volume[0][-1]
        volume[0] = [int(vol),int(no),date_seized,status,collection_count,link]
        
        for ii,articles in enumerate(volume[1]):
            del articles[2]
            del articles[0]
            articles[0] = articles[0].strip()
            authors = articles[1].replace(";",",")
            authors = authors.split(",")
            authors = [a.strip() for a in authors]
            authors = ",".join(authors)
            articles[1] = authors

            
# removing duplicate article listings
for k,v in unikorea_crawl.items():
    for volume in v:
        new_v = set(tuple(x) for x in volume[1])
        new_v = [list(x) for x in new_v]
        volume[1] = new_v

# removing index lists & empty lists
for k,v in unikorea_crawl.items():
    for i,volume in enumerate(v):
        info,articles = volume
        articles = [article for article in articles if "영문목차" not in article]
        articles = [article for article in articles if "편집부" not in article]
        volume[1] = [article for article in articles if "" not in article]
```

```python
# saving to json
import json
with open('unikorea_crawl.json', 'w') as f:
    json.dump(unikorea_crawl, f, ensure_ascii=False, indent=2)

# reloading for safety
import json
with open('unikorea_crawl.json',"r") as f:
    unikorea_crawl = json.load(f)
```

# Part II: NkResearch's collection relative to UniKorea's collection

```python
# importing master_dict for comparison
import json
with open('master_dict.json',"r") as f:
    master_dict = json.load(f)
```

```python
from operator import itemgetter

comparison = []

for subject,data in unikorea_crawl.items():
    for volume in data:
        ## finding journal match
        vol,no = volume[0][:2]
        hashes = []
        
        ## iterating master_dict for hashes
        for md5hash,metadata in master_dict.items():
            j = metadata["Journal Info"]
            if j["Name"] == subject:
                if j["Volume"] == vol:
                    if j["No"] == no:
                        hashes.append(md5hash)
        
        ## keying those hashes to compare information:
        nkresearch = []
        for h in hashes:
            title = master_dict[h]["Title"].strip()
            author = ",".join(master_dict[h]["Author"])
            start = master_dict[h]["Start Page"]
            end = master_dict[h]["End Page"]
            if not start:
                start = 0
                end = 0
            page_range = f"{start:03}"+"-"+ f"{end:03}"
            nkresearch.append([title,author,page_range])
        
        nkresearch = sorted(nkresearch, key=itemgetter(2))
        unikorea = sorted(volume[1], key=itemgetter(2))
        
        comparison.append([subject,vol,no,nkresearch,unikorea])
```

```python
from itertools import zip_longest  
import pandas.io.formats.excel
import pandas as pd
import xlsxwriter

row_count = 0
with xlsxwriter.Workbook('unikorea_vs_nkresearch.xlsx') as workbook:
    worksheet = workbook.add_worksheet()
    
    data_format1 = workbook.add_format({'bg_color': '#a3c9ff'})
    data_format2 = workbook.add_format({'bg_color': '#d1e4ff'})

    for journal_no, journal in enumerate(comparison):
        info = journal[0]+" Vol."+str(journal[1])+" No."+str(journal[2])
        
        nkr,uni = journal[-2:]
        nkr = sorted(nkr, key=itemgetter(1))
        uni = sorted(uni, key=itemgetter(1))
        zipped = list(zip_longest(nkr, uni,fillvalue=["","",""]))
    
        titles =[[item[0] for item in order] for order in zipped]
        authors = [[item[1] for item in order] for order in zipped]
        pages = [[item[2] for item in order] for order in zipped]
        
        idx = [i for i,title in enumerate(titles) if not title[1].isdigit()]
        
        titles = [title for i,title in enumerate(titles) if i in idx]
        authors = [author for i,author in enumerate(authors) if i in idx]
        pages = [page for i,page in enumerate(pages) if i in idx]
        
        
        worksheet.set_row(row_count, cell_format=data_format1)
        worksheet.write(row_count, 0, info)
        row_count+=1
        
        worksheet.set_row(row_count, cell_format=data_format2)
        worksheet.write(row_count, 0, "NkResearch Title")
        worksheet.write(row_count, 1, "UniKorea Title")
        for i,title in enumerate(titles):
            worksheet.write_row(row_count+i+1, 0, title)
            worksheet.set_column(0, 1, 50)
            
        worksheet.write(row_count, 2, "NkResearch Author")
        worksheet.write(row_count, 3, "UniKorea Author")
        for i,author in enumerate(authors):
            worksheet.write_row(row_count+i+1, 2, author)
            worksheet.set_column(2, 3, 10)

        worksheet.write(row_count, 4, "NkResearch Pages")
        worksheet.write(row_count, 5, "UniKorea Pages")
        for i,page in enumerate(pages):
            worksheet.write_row(row_count+i+1, 4, page)
            worksheet.set_column(4, 5, 10)
            
        row_count+=len(titles)
        row_count+=2
```

```python
# total entries
nkr_journal_count = 0
uni_journal_count = 0

nkr_article_count = 0
uni_article_count = 0
for collection in comparison:
    if collection[-2]:
        nkr_journal_count+=1
    if collection[-1]:
        uni_journal_count+=1
    nkr_article_count+=len(collection[-2])
    uni_article_count+=len(collection[-1])
df = pd.DataFrame([["Journal Count", nkr_journal_count,uni_journal_count],["Article Count",nkr_article_count,uni_journal_count]],columns=["",'NkResearch', 'UniKorea'])
df = df.style.set_table_attributes('style="font-size: 15px"')
df = df.hide_index()
df
```

```python
# finding missing journal entries
total_missing = 0
missing_journals = []
for journal_no, journal in enumerate(comparison):
    vol = journal[1]
    no = journal[2]
    nkr,uni = journal[-2:]
    if uni and not nkr:
        missing_journals.append([journal[0],vol,no,len(uni)])
        total_missing+=len(uni)
```

```python
def make_red(var):
    return "\033[31m"+str(var)+"\033[0m"
def make_bold(var):
    return "\033[01m"+str(var)+"\033[0m"

print ("\033[01mNkResearch\033[0m is missing a total of {} issues and {} articles relative to \033[01mUniKorea\033[0m".format(make_bold(make_red(len(missing_journals))),make_bold(make_red(total_missing))))
```

```python
missing_info = []
for journal in missing_journals:
    volume = journal[1:3]
    vol,no = volume
    article_count = journal[-1]
    for issue in unikorea_crawl[journal[0]]:
        info,articles = issue
        if volume == info[:2]:
            date,status,count,link = info[2:]
            missing_info.append([journal[0],vol,no,date,status,article_count,link])
```

```python
print (make_red("\033[01mNkResearch's repository is missing: "))
print ("---------------------------------")
for journal in missing_info:
    subject,vol,no,date,status,count,href = journal
    count = make_red(count)
    date = make_red(date)
    info_str = "Vol."+str(vol)+" No."+str(no)
    print ("{} {} with {} articles seized on {}".format(subject,info_str,count,date))

# making excel sheet with the same info
pandas.io.formats.excel.header_style = None
df = pd.DataFrame(missing_info)
writer = pd.ExcelWriter('nkresearch_missing.xlsx', engine='xlsxwriter')
df.to_excel(writer, 'Sheet1',index=False)
workbook = writer.book
worksheet = writer.sheets['Sheet1']

header_format = workbook.add_format({
        'bold': True,
        'text_wrap': True,
        'valign': 'top',
        'fg_color': '#d5fdfa',
        'border': 1})

header_info = ["Journal Subject","Volume","No","Date of Seizure","Status","Article Count","Link"]
for col_num, value in enumerate(df.columns.values):
    worksheet.write(0, col_num, header_info[col_num], header_format)
writer.save()
```

## 1. By Subject

```python
# by subject

unikorea_subjects = [subject for subject in unikorea_crawl]
nkresearch_subjects = [metadata["Journal Info"]["Name"] for md5hash,metadata in master_dict.items()]

unikorea_subjects = list(set(unikorea_subjects))
nkresearch_subjects = list(set(nkresearch_subjects))

unikorea_subject_stats = {}
for subject in unikorea_subjects:
    journal_count = len(unikorea_crawl[subject])
    article_count = [len(v[-1]) for v in unikorea_crawl[subject]]
    article_count = sum(article_count)
    unikorea_subject_stats[subject] = [journal_count,article_count]

nkresearch_subject_stats = {}
for subject in nkresearch_subjects:
    article_count = [metadata["Journal Info"] for md5hash,metadata in master_dict.items() if metadata["Journal Info"]["Name"] == subject]
    journal_count = len([dict(t) for t in {tuple(d.items()) for d in article_count}])
    article_count = len(article_count)
    nkresearch_subject_stats[subject] = [journal_count,article_count]

all_subjects = list(set(unikorea_subjects+nkresearch_subjects))

all_subject_stats = {}
for subject in all_subjects:
    if subject not in unikorea_subject_stats:
        all_subject_stats[subject] = [nkresearch_subject_stats[subject],[0,0]]
    else:
        stats = [nkresearch_subject_stats[subject],unikorea_subject_stats[subject]]
        all_subject_stats[subject] = stats
```

## 2. By Year

```python
unikorea_years = [subject for subject in unikorea_crawl]
nkresearch_subjects = [metadata["Journal Info"]["Name"] for md5hash,metadata in master_dict.items()]

unikorea_subjects = list(set(unikorea_subjects))
nkresearch_subjects = list(set(nkresearch_subjects))
```

```python
unikorea_vols = []
for subject,volumes in unikorea_crawl.items():
    unikorea_vols.append([volume[0][0] for volume in volumes])
unikorea_vols = [item for sublist in unikorea_vols for item in sublist]
unikorea_vols = list(set(unikorea_vols))

# since UniKorea does not explicitly state the year, we must deduce the year.
# one volume released per year; volume+1954 = year of release
# 1954 is the year after the Korean War ended
unikorea_years = sorted([vol+1954 for vol in unikorea_vols])
print ("Found years {} for UniKorea".format(unikorea_years))

nkresearch_years = [metadata["Journal Info"]["Year"] for md5hash,metadata in master_dict.items()]
nkresearch_years = sorted(list(set(nkresearch_years)))
print ("Found years {} for NkResearch\n".format(nkresearch_years))

all_years = sorted(list(set(unikorea_years+nkresearch_years)))
print ("All publication years:",all_years)
```

```python
# subject count, journal count, article count

unikorea = {}
for year in unikorea_years:
    unikorea_subjects = []
    unikorea_articles = []
    for subject,volumes in unikorea_crawl.items():
        for volume in volumes:
            if volume[0][0] == year-1954:
                unikorea_subjects.append(subject)
                unikorea_articles.append(len(volume[1]))
    unikorea_vols = len(unikorea_subjects)
    unikorea_articles = sum(unikorea_articles)
    unikorea_subjects = list(set(unikorea_subjects))
    unikorea[year] = [unikorea_subjects,unikorea_vols,unikorea_articles]

nkresearch = {}
for year in nkresearch_years:
    nkresearch_articles = [metadata["Journal Info"] for md5hash,metadata in master_dict.items() if metadata["Journal Info"]["Year"] == year]
    nkresearch_subjects = [m["Name"] for m in nkresearch_articles]
    nkresearch_subjects = list(set(nkresearch_subjects))
    nkresearch_vols = len([dict(t) for t in {tuple(d.items()) for d in nkresearch_articles}])
    nkresearch_articles = len(nkresearch_articles)
    nkresearch[year] = [nkresearch_subjects,nkresearch_vols,nkresearch_articles]
```

```python
# displaying subjects first

color = ['#ffd6df','#ffdede','#ffdad1','#ffe8d5','#ffeecc','#ffffcc','#f8ffde','#e3ffd9','#d9fff2','#d9fbff','#d9f6ff','#d9e7ff','#fdf0db','#dcd9ff','#e8d9ff','#fed9ff']
color_dict = dict(zip(sorted(all_subjects)[::-1],color))

nkresearch_subject = {}
unikorea_subject = {}

nkresearch_subjects

for year in all_years:
    if year in nkresearch:
        nkresearch_subject[year] = nkresearch[year][0]
    else:
        nkresearch_subject[year] = []
    if year in unikorea:
        unikorea_subject[year] = unikorea[year][0]
    else:
        unikorea_subject[year] = []

fig, (ax1, ax2) = plt.subplots(2,sharex=True,figsize=(11, 7))
fig.suptitle('Collected Journal Subjects by Year', x=0.5, y=0.97,size=15)

x = all_years

ax1.ticklabel_format(useOffset=False)
ax2.ticklabel_format(useOffset=False)

ax1.set_xlim(min(x)-0.6, max(x)+0.8)
ax1.set_ylim(-1, 14)

ax2.set_xlim(min(x)-0.6, max(x)+0.8)
ax2.set_ylim(-1, 14)

ax1.set_xticks(np.arange(min(x), max(x)+0.8, 1.0))
ax1.set_yticks(np.arange(0, 14, 1.0))

ax2.set_xticks(np.arange(min(x), max(x)+0.8, 1.0))
ax2.set_yticks(np.arange(0, 14, 1.0))

                  
for i,(year,subjects) in enumerate(nkresearch_subject.items()):
    ax1.annotate(str(len(subjects)), (x[i], len(subjects)+0.8),size=8.5,ha='center',va='top')
    subjects = sorted(subjects)
    for ii,subject in enumerate(subjects):
        bbox = dict(boxstyle='square,pad=0.15', facecolor=color_dict[subject], alpha=0.95)
        ax1.annotate(subject, (year, ii),size=6.5,ha='center',va='bottom',bbox=bbox)


plt.subplots_adjust(wspace=3, hspace=0.1)
for i,(year,subjects) in enumerate(unikorea_subject.items()):
    ax2.annotate(str(len(subjects)), (x[i], len(subjects)+0.8),size=8.5,ha='center',va='top')
    subjects = sorted(subjects)
    for ii,subject in enumerate(subjects):
        bbox = dict(boxstyle='square,pad=0.15', facecolor=color_dict[subject], alpha=0.95)
        ax2.annotate(subject, (year, ii),size=6.5,ha='center',va='bottom',bbox=bbox)
        
ax1.plot(x,[len(subjects)-0.05 for year,subjects in nkresearch_subject.items()],linewidth=1,linestyle='dotted',color='tab:red',alpha=0.8)
ax1.fill_between(x, [len(subjects)-0.05 for year,subjects in nkresearch_subject.items()],alpha=0.1,color='tab:red')
ax1.legend(loc='upper left', labels=['NkResearch'],frameon=True,labelcolor='linecolor')

ax2.plot(x,[len(subjects)-0.05 for year,subjects in unikorea_subject.items()],linewidth=1,linestyle='dotted',color='tab:blue',alpha=0.8)
ax2.fill_between(x, [len(subjects)-0.05 for year,subjects in unikorea_subject.items()],alpha=0.1,color='tab:blue')
ax2.legend(loc='upper left',labels=['UniKorea'],frameon=True,labelcolor='linecolor')

ax1.yaxis.set_tick_params(labelsize='small')
ax1.yaxis.label.set_color('tab:red')
ax1.spines['left'].set_color('tab:red')
ax1.spines['right'].set_color('tab:red')
ax1.spines['top'].set_color('tab:red')

ax2.yaxis.set_tick_params(labelsize='small')
ax2.yaxis.label.set_color('tab:blue')
ax2.spines['left'].set_color('tab:blue')
ax2.spines['right'].set_color('tab:blue')
ax2.spines['bottom'].set_color('tab:blue')


ax2.tick_params(labelbottom=False,labeltop=True)
ax2.tick_params(axis='x', color='none', pad=1.5,labelsize=13)

ax1.tick_params(axis='y', colors='tab:red')
ax2.tick_params(axis='y', colors='tab:blue')

ax1.tick_params(axis='y', color='tab:red', size=3)
ax2.tick_params(axis='y', color='tab:blue', size=3)

ax1.set_ylabel('Number of Subjects',ha='center',va='bottom',color="tab:red")
ax2.set_ylabel('Number of Subjects',ha='center',va='bottom',color="tab:blue")
fig.tight_layout()
plt.savefig('subjects_by_year.png',dpi=1200)

plt.show()
```
