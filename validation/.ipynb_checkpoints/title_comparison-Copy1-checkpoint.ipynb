# Comparing Titles with UniKorea's repository

```python
# load data
import json
with open('master_dict.json',"r") as f:
    master_dict = json.load(f)

import json
with open('unikorea_crawl.json',"r") as f:
    unikorea_crawl = json.load(f)
```

```python
import difflib
from operator import itemgetter

comparison = []

for subject,data in unikorea_crawl.items():
    for volume in data:
        ## finding journal match
        vol,no = volume[0][:2]
        hashes = []
        
        ## iterating master_dict for hashes
        for md5hash,metadata in master_dict.items():
            j = metadata["Journal Info"]
            if j["Name"] == subject:
                if j["Volume"] == vol:
                    if j["No"] == no:
                        hashes.append(md5hash)
        
        ## keying those hashes to compare information:
        nkresearch = []
        for h in hashes:
            title = master_dict[h]["Title"].strip()
            author = ",".join(master_dict[h]["Author"])
            nkresearch.append([title,author])
        
        nkresearch = sorted(nkresearch, key=itemgetter(0))
        unikorea = [article[:2] for article in volume[1]]
        unikorea = sorted(unikorea, key=itemgetter(0))
        
        comparison.append([subject,vol,no,nkresearch,unikorea])
```

```python
def levenshtein_sort(diff1,diff2):
    cands1 = [diff[0] for diff in diff1]
    cands2 = [diff[0] for diff in diff2]
    
    match_idx = []
    for i,cand1 in enumerate(cands1):
        distances= [levenshtein_distance(cand1,cand2) for cand2 in cands2]
        best_idx = distances.index(min(distances))
        match_idx.append(best_idx)
    duplicate = [idx for idx, item in enumerate(match_idx) if item in match_idx[:idx]]
    diff2_matched = [diff2[i] for i in match_idx]
    return diff1,diff2_matched
```

```python
def duplicate(diff1,diff2):
    cands1 = [diff[0] for diff in diff1]
    cands2 = [diff[0] for diff in diff2]
    
    match_idx = []
    for i,cand1 in enumerate(cands1):
        distances= [levenshtein_distance(cand1,cand2) for cand2 in cands2]
        best_idx = distances.index(min(distances))
        match_idx.append(best_idx)
    duplicate = [idx for idx, item in enumerate(match_idx) if item in match_idx[:idx]]
    if duplicate:
        return True
    else:
        return False
```

```python
def levenshtein_duplicate(diff1,diff2):
    cands1 = [diff[0] for diff in diff1]
    cands2 = [diff[0] for diff in diff2]
    
    match_idx = []
    for i,cand1 in enumerate(cands1):
        distances= [levenshtein_distance(cand1,cand2) for cand2 in cands2]
        best_idx = distances.index(min(distances))
        match_idx.append(best_idx)
    duplicate = [idx for idx,item in enumerate(match_idx) if item in match_idx[:idx]]
    diff1_duplicate = [diff1[i] for i in duplicate]
    duplicate_idx = [item for idx,item in enumerate(match_idx) if item in match_idx[:idx]]
    re_match_idx = [item for idx,item in enumerate(match_idx) if idx not in duplicate]
    diff2_matched = [diff2[i] for i in re_match_idx]
    re_diff1 = [d for d in diff1 if d not in diff1_duplicate]
    
    diff2_duplicate = [diff2[i] for i in duplicate_idx]
    
    diff2_unmatched = [d for d in diff2 if d not in diff2_matched]
    return re_diff1,diff2_matched,diff1_duplicate,diff2_unmatched

def duplicate(diff1,diff2):
    cands1 = [diff[0] for diff in diff1]
    cands2 = [diff[0] for diff in diff2]
    
    match_idx = []
    for i,cand1 in enumerate(cands1):
        distances= [levenshtein_distance(cand1,cand2) for cand2 in cands2]
        best_idx = distances.index(min(distances))
        match_idx.append(best_idx)
    duplicate = [idx for idx, item in enumerate(match_idx) if item in match_idx[:idx]]
    if duplicate:
        return True
    else:
        return False
```

```python
comparison2 = comparison
nkresearch_missing = []
unikorea_missing = []
comparison3 = []

for collection in comparison2:
    nkresearch,unikorea = collection[-2:]
    if nkresearch:
        nkresearch=sorted(nkresearch)
        unikorea=sorted(unikorea)
        
        if unikorea == nkresearch:
            comparison3.append(collection)
            continue
            
        elif len(unikorea) == len(nkresearch):
            a = nkresearch
            b = unikorea
            same = [n for n in a if n in b]
            nk_cand = [n for n in a if n not in b]
            un_cand = [u for u in b if u not in a]
            b2,a2 = levenshtein_sort(un_cand,nk_cand)
            a_unmatched = [aa for aa in nk_cand if aa not in a2]
            if len(a_unmatched) == 0:
                comparison3.append([collection[:3],same+a2,same+b2])
            elif duplicate(nkresearch,unikorea):
                a,b = nkresearch,unikorea
                n_matched,u_matched,n_unmatched,u_unmatched = levenshtein_duplicate(a,b)
                nkresearch_missing.append([collection[:3],u_unmatched])
                unikorea_missing.append([collection[:3],n_unmatched])
                comparison3.append([collection[:3],n_matched,u_matched])
            elif duplicate(unikorea,nkresearch):
                u_matched,n_matched,u_unmatched,n_unmatched = levenshtein_duplicate(b,a)
                nkresearch_missing.append([collection[:3],u_unmatched])
                unikorea_missing.append([collection[:3],n_unmatched])
                comparison3.append([collection[:3],n_matched,u_matched])
            else:
                print("what")
        
        elif len(nkresearch) > len(unikorea):
            a = nkresearch
            b = unikorea
            same = [n for n in a if n in b]
            nk_cand = [n for n in a if n not in b]
            un_cand = [u for u in b if u not in a]
            b2,a2 = levenshtein_sort(un_cand,nk_cand)
            a_unmatched = [aa for aa in nk_cand if aa not in a2]
            if len(a_unmatched) == len(nkresearch) - len(unikorea):
                comparison3.append([collection[:3],same+a2,same+b2])
                unikorea_missing.append([collection[:3],a_unmatched])
            elif duplicate(un_cand,nk_cand):
                u_matched,n_matched,u_unmatched,n_unmatched = levenshtein_duplicate(b,a)
                nkresearch_missing.append([collection[:3],u_unmatched])
                unikorea_missing.append([collection[:3],n_unmatched])
                comparison3.append([collection[:3],n_matched,u_matched])
            else:
                print("what")
        elif len(nkresearch) < len(unikorea):
            a = nkresearch
            b = unikorea
            nk_cand = [n for n in a if n not in b]
            un_cand = [u for u in b if u not in a]
            same = [n for n in a if n in b]
            a2,b2 = levenshtein_sort(nk_cand,un_cand)
            b_unmatched = [bb for bb in un_cand if bb not in b2]
            if len(b_unmatched) == len(unikorea) - len(nkresearch):
                comparison3.append([collection[:3],same+a2,same+b2])
                nkresearch_missing.append([collection[:3],b_unmatched])
            elif duplicate(nk_cand,un_cand):
                n_matched,u_matched,n_unmatched,u_unmatched = levenshtein_duplicate(a,b)
                nkresearch_missing.append([collection[:3],u_unmatched])
                unikorea_missing.append([collection[:3],n_unmatched])
                comparison3.append([collection[:3],n_matched,u_matched])
            else:
                nkresearch_missing.append([collection[:3],b_unmatched])
```

```python
def make_red(var):
    return "\033[31m"+str(var)+"\033[0m"

def make_blue(var):
    return "\033[34m"+str(var)+"\033[0m"

def make_green(var):
    return "\033[32m"+str(var)+"\033[0m"

def make_bold(var):
    return "\033[1m"+str(var)+"\033[0m"

def diff_strings(a, b):
    output = []
    matcher = difflib.SequenceMatcher(None, a, b)
    for opcode, a0, a1, b0, b1 in matcher.get_opcodes():
        if opcode == "equal":
            output.append(a[a0:a1])
        elif opcode == "insert":
            output.append(make_green(b[b0:b1]))
        elif opcode == "delete":
            output.append(make_red(a[a0:a1]))
        elif opcode == "replace":
            output.append(make_green(b[b0:b1]))
            output.append(make_red(a[a0:a1]))
    return "".join(output)
```

```python
lev_distances = []
for collection in comparison3:
    nkresearch,unikorea = collection[-2:]
    for i,article in enumerate(nkresearch):
        lev_dist = levenshtein_distance(article[0],unikorea[i][0])
        lev_distances.append(lev_dist)
print ("Average Levenshtein distance is", sum(lev_distances)/len(lev_distances))
```

```python
import numpy as np
a = np.array(lev_distances)

percentiles_low = {}
for v in list(range(0,10)):
    p = np.percentile(a, v)
    p_str = str(v)+"th"
    percentiles_low[p_str] = p
percentiles_high = {}
for v in list(range(90,100)):
    p = np.percentile(a, v)
    p_str = str(v)+"th"
    percentiles_high[p_str] = p
df1 = pd.DataFrame(percentiles_low, index=['Levenshtein Distance'])
df1 = df1.style.set_table_attributes('style="font-size: 12px"')
df2 = pd.DataFrame(percentiles_high, index=['Levenshtein Distance'])
df2 = df2.style.set_table_attributes('style="font-size: 12px"')
df1
```
