```python
# reloading for safety
import json
with open('aligned.json',"r") as f:
    aligned_data = json.load(f)
```

```python
import pandas as pd

multiindex_labels = []
for journal in aligned_data:
    journal_tuple = (journal[0][0],journal[0][1],journal[0][2])
    multiindex_labels.append(journal_tuple)
multiindex_labels[:4]
```

```python
dfs = []
for journal in aligned_data:
    nkr,uni = journal[-2:]
    df_nkr=pd.DataFrame({'Title':[n[0] for n in nkr],'Author':[n[1] for n in nkr]})
    df_uni=pd.DataFrame({'Title':[u[0] for u in uni],'Author':[u[1] for u in uni]})
    df = pd.concat([df_nkr,df_uni],keys=['NkResearch','UniKorea'])
    dfs.append(df)
```

```python
aligned_df = pd.concat(dfs,keys=multiindex_labels)
aligned_df
```

```python
import xlsxwriter

writer = pd.ExcelWriter('aligned.xlsx', engine='xlsxwriter')

aligned_df.to_excel(writer, sheet_name='Sheet1',engine='xlsxwriter')
writer.save()
```
